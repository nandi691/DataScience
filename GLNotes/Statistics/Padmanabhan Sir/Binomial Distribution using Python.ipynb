#  Probability Distributions

Probability distributions are used to understand and analyze the patterns in the data. It is a mathematical function that gives the probabilities of all the outcomes of an experiment. 
There are two types of probability distributions based on the type of random variable:
     <li> Discrete probability distributions and
     <li> Continuous probability distributions

##  Discrete Probability Distributions

It is the probability distribution related to the discrete random variable. In this case, the `probability mass function (pmf)` gives the probability that a variable is equal to a specific value. For a discrete random variable `X`, pmf of X is given as $ P(X = x_{i})$, where $x_{i}$ for i = 1, 2, ..., n are the values taken by the variable X.

The sum of all the probabilities is always 1. i.e. $\sum_{i = 1}^{n} P(X = x_{i}) = 1$

For a discrete random variable X with values $x_{1}, x_{2}, ..., x_{n}$ the `cumulative distribution function (cdf)` is denoted by F(x) and defined as:

<p style='text-indent:25em'> <strong> $ F(x_{i}) = P[X \leq x_{i}]$</strong> </p>

The cdf of a random variable returns the probability that the variable takes all the values less than or equal to the specific value.

# Binomial Distribution

#### The binomial distribution is one of the most commonly used distributions in statistics. It describes the probability of obtaining 'k' successes in 'n' binomial experiments.
A discrete variable X taking values 0, 1, 2,..., n follows a binomial distribution with parameters `n` and `p`, if the pmf of X is given by:

<p style='text-indent:25em'> <strong> $ P(X = k) = {n \choose k} p^{k}q^{n-k}$</strong> $\hspace{2cm}$  k = 0, 1, ..., n </p>

Here, `p` denotes the probability of success  and `q` denotes the probability of failure of a random experiment such that p + q = 1

If `n` independent bernoulli trials (each with `p` as probability of success) are executed then the number of successes follows a binomial distribution.


#### If a random variable X follows a binomial distribution, then the probability that X = k successes can be found by using the PYTHON syntax:


#### Note:

(1) Probability of exactly 'k' successes = P(X = k)

(2) Probability of at most 'k' successes = P(X $\leq$ k)

(3) Probability of at least 'k' successes = P(X $\geq$ k) = 1 - P(X < k)

(4) Probability of not more than 'k' successes = P(x $\leq$ k)

## How to Generate a Binomial Distribution
You can generate an array of values that follow a binomial distribution by using the random.binomial function from the numpy library:

```python
from numpy import random

#generate an array of 10 values that follow a binomial distribution

random.binomial(n=10, p=.25, size=10)

```

Each number in the resulting array represents the number of successes experienced during 10 trials where the probability of success in a given trial was .25.

## How to Calculate Probabilities Using a Binomial Distribution
### You can use the binom.pmf(k, n, p),  binom.cdf(k, n, p) and binom.sf(k, n, p) functions to calculate probabilities related to the Binomial distribution.

binom.pmf(k, n, p): This is the Probability Mass Function (PMF). It calculates the probability of getting exactly k successes in n trials, where the probability of success on a single trial is p.

binom.cdf(k, n, p): This is the Cumulative Distribution Function (CDF). It calculates the probability of getting k successes or fewer in n trials, where the probability of success on a single trial is p.

binom.sf(k, n, p): This is the Survival Function, also defined as 1 - cdf. It calculates the probability of getting more than k successes in n trials, where the probability of success on a single trial is p. Its sometimes more accurate than 1 - cdf.

### Syntax: 

    
#### (1) The probability of 'k' successes [i.e., P(X = k)] out of 'n' trails is given by
       binom.pmf(k, n, p) 
#### (2) The probability of at most 'k' successes [i.e., P(X $\leq$ k)] out of 'n' trails is given by
       binom.cdf(k, n, p) 
#### (3) The probability of more than 'k' successes [i.e., P(X > k)] out of 'n' trails is given by
       1 - binom.cdf(k, n, p) 
             (or)
           binom.sf(k,n,p)
#### (4) The probability of less than 'k' successes [i.e., P(X < k)] out of 'n' trails is given by
       binom.cdf(k-1, n, p)
#### (5) The probability of at least 'k' successes [i.e., P(X $\geq$ k)] out of 'n' trails is given by
       1 - binom.cdf(k-1, n, p)    
#### (6) The probability of success between  'k1' and 'k2' (k1 > k2) out of 'n' trails is given by
       binom.cdf(k1, n, p) - binom.cdf(k2, n, p)  

### Question 1: Nathan makes 60% of his free-throw attempts. If he shoots 12 free throws, what is the probability that he makes exactly 10?

```python
from scipy.stats import *

# Given data
n = 12    # number of free throws
p = 0.6   # probability of success = 60% of his free-throw attempts
k = 10    # makes exactly 10

#calculate binomial probability

probability=binom.pmf(k, n, p)

print('The probability that Nathan makes exactly 10 free throws is',round(probability,4))
```

### Question 2: In a shooting academy, data was collected on the precision shooting of a student. From 15 shots fired 11 were on target. Consider the same student, what is the probability that out of 50 shots fired, exactly 35 will hit the target?

```python
from scipy.stats import *

# Given data
n = 50      # out of 50 shots fired
p = 11/15   # probability of success =  11 shots fired from 15 shots
k = 35      # exactly 35 will hit the target

#calculate binomial probability

probability=binom.pmf(k, n, p)

print('The probability that out of 50 shots fired, exactly 35 will hit the target is',round(probability,3))
```

### Question 3: Marty flips a fair coin 5 times. What is the probability that the coin lands on heads 2 times or fewer?

```python
from scipy.stats import *

# Given data
n = 5       # flips a fair coin 5 times
p = 0.5     # probaility of getting head when tossing a coin 
k = 2       # coin lands on heads 2 times or fewer (less than or equal to 2)

#calculate binomial probability

probability=binom.cdf(k, n, p)

print('The probability that the coin lands on heads 2 times or fewer is',round(probability,3))
```

### Question 4: Find the probability of at most two heads, when three coins are tossed simultaneously. 



```python
from scipy.stats import *

# Given data
n = 3       # flips a fair coin 5 times
p = 0.5     # probaility of getting head when tossing a coin 
k = 2       # atmost two heads (less than or equal to 2)

#calculate binomial probability

probability=binom.cdf(k, n, p)

print('The probability of at most two heads is',round(probability,3))
```

### Question 5: If three coins are tossed simultaneously, find the probability of getting at least two heads.


```python
from scipy.stats import *

# Given data
n = 3       # flips a fair coin 5 times
p = 0.5     # probaility of getting head when tossing a coin 
k = 2       # at least two heads (greater than or equal to 2)

#calculate binomial probability

probability=1-binom.cdf(k-1, n, p)

print('The probability of getting at least two heads is',round(probability,3))
```

### Question 6: It is known that 70% of individuals support a certain law. If 10 individuals are randomly selected, what is the probability that between 4 and 6 of them support the law?

```python
from scipy.stats import *

# Given data
n = 10                # 10 individuals are randomly selected
p = 0.7               # 70% of individuals support a certain law
k1, k2 = 6, 4         # between 4 and 6 of them support the law

#calculate binomial probability

probability=binom.cdf(k1, n, p)-binom.cdf(k2, n, p)

print('The probability that between 4 and 6 of the randomly selected individuals support the law is',round(probability,3))
```

### Question 7: Heaven Furnitures (HF) sells furniture like sofas, beds and tables. It is observed that 25% of their customers complain about the furniture purchased by them for many reasons. On Tuesday, 20 customers purchased furniture products from HF. Calculate the probability that (a)  exactly 3 customers will complain about the purchased products, (b) more than 3 customers will complain about the furniture purchased by them.

```python
from scipy.stats import *

# Given data
n = 20      # 20 customers purchased furniture
p = 0.25   # probability of success = 25% of their customers complain
k = 3     # exactly 3 customers will complain

print('(a) The probability of exactly 3 customers will complain about the purchased products is',round(binom.pmf(k, n, p),3))

print('(b) The probability of more than 3 customers will complain about the purchased products is',round(binom.sf(k, n, p),3))
```

### Question 8: The probability that a man aged 60 will live upto 70 is 0.65. Out of 10 men, now at the age of 60, find the probability that (i) at least 7 will live upto 70, (ii) exactly 9 will live upto 70, (iii) at most 9 will live upto 70.

```python
from scipy.stats import *

# Given data
n = 10      # 12 files are opened in an hour
p = 0.65   # man aged 60 will live upto 70 is 0.65

print('(i) The probability of at least 7 will live upto 70 is',round(1-binom.cdf(6, n, p),3))  

print('(ii) The probability of exactly exactly 9 will live upto 70 is',round(binom.pmf(9, n, p),3))

print('(iii) The probability of at most 9 will live upto 70 is',round(binom.cdf(9, n, p),3))
```

### Question 9: When a coin is tossed 4 times, find the probability of getting (i) exactly one head, (ii) at most 3 heads and (iii) at least two heads.

```python
from scipy.stats import *

# Given data
n = 4      # coin is tossed 4 times
p = 1/2   # probability of getting head

print('(i) The probability of getting exactly one head is',round(binom.pmf(1, n, p),4))  

print('(ii) The probability of getting at most 3 heads is',round(binom.cdf(3, n, p),4))

print('(iii) The probability of getting at least two heads is',round(1-binom.cdf(1, n, p),4))
```

### Question 11: Out of 800 families with 5 children each, how many would you expect to have (i) 3 boys, (ii) 5 girls, (iii) either 2 or 3 boys? (Assume equal probabilities for boys and girls)

```python
from scipy.stats import *

# Given data
n = 5      # 5 children each in 800 families
p = 1/2   # probability of boy

print('(i) From 800 families,the probability of number of families having 3 boys is', round(binom.pmf(3, n, p)*800))

print('(ii) From 800 families,the probability of number of families having 5 girls is', round(binom.pmf(0, n, p)*800)) # no boys

res=binom.pmf(2, n, p) + binom.pmf(3, n, p) # either 2 or 3 boys

print('(iii) From 800 families,the probability of number of families having either 2 or 3 boys is', round(res*800))
```

### Question 12: The probability that a pen manufactured by a company will be defective is 1/10 . If 12 such pens are manufactured, find the probability that (i) exactly two will be defective, (ii) at least two will be defective, (iii) none will be defective.

## How to Visualize a Binomial Distribution

You can visualize a binomial distribution in Python by using the seaborn and matplotlib libraries:

### Find the probability distribution of getting a head when tossing a coin thrice and show it graphically?

```python
from scipy.stats import *
n=3
p=0.5
x=np.arange(0,4)
print('The probability distribution of getting a head when tossing a coin thrice is\n\n ',binom.pmf(x,n,p))
pmf=binom.pmf(x,n,p)
plt.bar(x,pmf)
plt.xlabel('Head', fontsize = 15)
plt.ylabel('Probability', fontsize = 15)
plt.xticks([0,1,2,3,4])
plt.title('Probability Mass Function of X', fontsize = 15)

plt.show()
```
