##  Continuous Probability Distributions

It is the probability distribution related to the continuous random variable. The area under `probability density function (pdf)` gives the probability that a variable lies in a specific range. The value of pdf at a specific point is always 0.

Probability that X lies between [a,b] is given by:

<p style='text-indent:25em'> <strong> $P(a \leq X \leq b) = \int_{a}^{b} f(x) dx$</strong> </p> 

Area under the curve is 1. i.e. $\int_{-\infty}^{\infty} f(x) dx = 1$

For a continuous random variable X the `cumulative distribution function (cdf)` is denoted by F(x) and defined as:

<p style='text-indent:25em'> <strong> $ F(x) = \int_{-\infty}^{x} f(u) du$</strong> </p>

The cdf of a random variable returns the probability that the variable takes all the values less than or equal to the specific value.

# Normal Distribution

A continuous variable X follows a normal distribution with the parameters $\mu$ and $\sigma$ if the pdf of X is given by:

<p style='text-indent:25em'> <strong> $f(x) = \frac {1}{\sigma\sqrt(2 \pi)} e^\frac {-(x-\mu^2)}{2\sigma^2}$</strong></p>

The graph of the normal distribution is called the normal curve. It is bell-shaped and symmetrical about the line x=$\mu$. The line   divides the total area under the curve which is equal to 1 into two equal parts. The area to the right as well as to the left of the line x=$\mu$  is 0.5.



The normal distribution is a bell-shaped symmetric distribution with `mean = median = mode`. 

The mean and variance of the distribution is given as:<br>

Mean = $\mu$ 

Variance = $\sigma^{2}$ 



### Example:

## Standard Normal Distribution


## The variable Z given as, Z = $\frac{X - \mu}{\sigma}$ follows a `standard normal distribution` with mean 0 and variance 1. The random variable Z is called the standard normal random variable.

## How to Calculate Probabilities Using a Normal Distribution
### You can use the norm.pdf(z),  norm.cdf(z) and norm.sf(z) functions to calculate probabilities related to the Normal distribution.

### Syntax: Z = $\frac{X - \mu}{\sigma}$

    
#### (1) The probability of 'k' successes [i.e., P(X = k)] out of 'n' trails is given by
       norm.pdf(z) 
#### (2) The probability of at most 'k' successes [i.e., P(X $\leq$ k)] out of 'n' trails is given by
       norm.cdf(z) 
#### (3) The probability of more than 'k' successes [i.e., P(X > k)] out of 'n' trails is given by
       1 - norm.cdf(z) 
           OR
       norm.sf(z)
#### (4) The probability of less than 'k' successes [i.e., P(X < k)] out of 'n' trails is given by
       norm.cdf(z)
#### (5) The probability of at least 'k' successes [i.e., P(X$\geq$ k)] out of 'n' trails is given by
       1 - norm.cdf(z)  
           OR
       norm.sf(z)
#### (6) The probability of success between  'k1' and 'k2' (k1 > k2) out of 'n' trails is given by
       norm.cdf(z1) - norm.cdf(z2)  

### Question 1: A survey was conducted and it was found that the people spend their 300 minutes in a day surfing on online shopping sites on average and the corresponding standard deviation is 127 minutes. Assume that the time spent on surfing follows a normal distribution. Calculate the following probabilities:
#### a. What is the probability that the users are spending less than or equal to 100 minutes per day?
#### b. What is the probability that people are spending more than 400 minutes on online shopping sites per day?
#### c. What is the probability that people are spending time between 250 minutes and 350 minutes per day?

```python
from scipy.stats import *

# Given data
avg = 300     # on average people spend 300 minutes everyday 
std = 127     # standard deviation is 127 minutes

z1 = (100 - avg) / std # standardize the variable with x = 100
z2 = (400 - avg) / std # standardize the variable with x = 400
z3 = (350 - avg) / std # standardize the variable with x = 350
z4 = (250 - avg) / std # standardize the variable with x = 250

prob1 = round(norm.cdf(z1),2)
#prob2 = round(1-norm.cdf(z2),2)
prob2 = round(norm.sf(z2),2)
prob3 = round(norm.cdf(z3)-norm.cdf(z4),2)

print('a. The probability that the users are spending less than or equal to 100 minutes daily is', prob1)
print('b. The probability that people are spending more than 400 minutes on online shopping sites per day is', prob2)
print('c. The probability that people are spending time between 250 minutes and 350 minutes per day is', prob3)
```

### Question 2: A monthly balance in the bank account of a credit card holders is assumed to be normally distributed with mean 500 dollars and variance 100 dollars. What is the probability that the balance can be more than 513.5 dollars?

```python
from scipy.stats import *
import math
# Given data

avg = 500                     # average account balance
var = 100                    # variance is 100 dollars
std = math.sqrt(var)           # standard deviation is square-root of variance 

z = (513.5 - avg) / std     # standardize the variable with x = 513.5

# prob = round(1-norm.cdf(z),3)
prob = round(norm.sf(z),3)

print('The probability that the balance can be more than 513.5 dollars is', prob)
```

### Question 3: In a test on 2000 electric bulbs, it was found that the life of a particular make was normally distributed with an average life of 2040 hours and S.D. of 60 hours. Estimate the number of bulbs likely to burn for (i) more than 2150 hours, (ii) less than 1950 hours, (iii) more than 1920 hours and less than 2160 hours.

```python
from scipy.stats import *

# Given data:   test on 2000 electric bulbs
avg = 2040     # average life of 2040 hours
std = 60      # S.D. of 60 hours

z1 = (2150 - avg) / std             # standardize the variable with x = 2150
z2 = (1950 - avg) / std             # standardize the variable with x = 1950
z3 = (2160 - avg) / std             # standardize the variable with x = 2160
z4 = (1920 - avg) / std             # standardize the variable with x = 1920

prob1 = round(1-norm.cdf(z1),4)
prob2 = round(norm.cdf(z2),4)
prob3 = round(norm.cdf(z3)-norm.cdf(z4),4)

print('(i) The number of bulbs likely to burn for more than 2150 hours =', round(prob1*2000))
print('(ii) The number of bulbs likely to burn for less than 1950 hours =', round(prob2*2000))
print('(iii) The number of bulbs likely to burn for more than 1920 hours and less than 2160 hours is', round(prob3*2000))
```

### Question 6: In an examination taken by 500 candidates, the average and the standard deviation of marks obtained (normally distributed) are 40% and 10% respectively. Find approximately (i) how many will pass, if 50% is fixed as a minimum  and (ii) how many have scored marks above 60%?

```python
from scipy.stats import *

# Given data:  examination taken by 500 candidates
avg = 40    
std = 10      

z1 = (50 - avg) / std             # standardize the variable with x = 50
z2 = (60 - avg) / std             # standardize the variable with x = 60

prob1 = round(1-norm.cdf(z1),4)
prob2 = round(1-norm.cdf(z2),4)

print('(i) The number of students will pass, if 50% is fixed as a minimum =', round((prob1*500)))
print('(ii) The number of students have scored marks above 60% = ', round((prob2*500)))
```

### Question 7:The annual salaries of employees in a large company are approximately normally distributed with a mean of 50,000 and a standard deviation of 20,000. (а) What percent of people earn less than 40,000? (b) What percent of people earn between 45,000 and 65,000? (c) What percent of people earn more than 70,000?
                                                          NOV 2023

```python
from scipy.stats import *

# Define the mean and standard deviation
mean = 50000
std_dev = 20000

# Calculate the Z-scores
z1 = (40000 - mean) / std_dev
z2 = (45000 - mean) / std_dev
z3 = (65000 - mean) / std_dev
z4 = (70000 - mean) / std_dev

# Calculate the percentages
p1 = norm.cdf(z1) * 100
p2 = (norm.cdf(z3) - norm.cdf(z2)) * 100
#p3 = (1 - norm.cdf(z3)) * 100
p3 = norm.sf(z4) * 100

# Print the results
print(f"(a) The percentage of people who earn less than 40,000 is {p1:.2f}%")
print(f"(b) The percentage of people who earn between 45,000 and 65,000 is {p2:.2f}%")
print(f"(c) The percentage of people who earn more than 70,000 is {p3:.2f}%")
```
