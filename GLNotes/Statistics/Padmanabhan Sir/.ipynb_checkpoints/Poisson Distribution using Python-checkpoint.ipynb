# Poisson Distribution

In a binomial distribution, if the following axioms hold:

(i)	 the number of trails n is large, 
(ii) the probability of success is closer to zero and 
(iii) np = $\lambda$ is a finite number 

then the probability distribution reduces to 
<p style='text-indent:25em'> <strong> $ P(X = k) = \frac{e^{-\lambda}\lambda^{k}}{k!}$</strong> $\hspace{2cm}$  k = 0, 1, 2,... </p>

This probability distribution is called Poisson distribution.

**Note:** Consider a variable X that follows a binomial distribution with parameters `n` and `p`. If n$\rightarrow$$\infty$ and p$\rightarrow$0 then X follows a poisson distribution with parameter $\lambda$ = np`.




#### Note:

(1) Probability of exactly 'k' successes = P(X = k)

(2) Probability of at most 'k' successes = P(X $\leq$ k)

(3) Probability of at least 'k' successes = P(X $\geq$ k) = 1 - P(X < k)

(4) Probability of not more than 'k' successes = P(x $\leq$ k)

## How to Calculate Probabilities Using a Poisson Distribution
### You can use the poisson.pmf(k, mu),  poisson.cdf(k, mu) and poisson.sf(k, mu) functions to calculate probabilities related to the Poisson distribution.

### Syntax: 

    
#### (1) The probability of 'k' successes [i.e., P(X = k)] out of 'n' trails is given by
       poisson.pmf(k, mu) 
#### (2) The probability of at most 'k' successes [i.e., P(X $\leq$ k)] out of 'n' trails is given by
       poisson.cdf(k, mu) 
#### (3) The probability of more than 'k' successes [i.e., P(X > k)] out of 'n' trails is given by
       1 - poisson.cdf(k, mu) 
            OR
       poisson.sf(k, mu)
#### (4) The probability of less than 'k' successes [i.e., P(X < k)] out of 'n' trails is given by
       poisson.cdf(k-1, mu)
#### (5) The probability of at least 'k' successes [i.e., P(X$\geq$ k)] out of 'n' trails is given by
       1 - poisson.cdf(k-1, mu)    
#### (6) The probability of success between  'k1' and 'k2' (k1 > k2) out of 'n' trails is given by
       poisson.cdf(k1, mu) - poisson.cdf(k2, mu)  

#### Question 1: A store sells 3 apples per day on average. What is the probability that they will sell 5 apples on a given day? 

```python
from scipy.stats import *

# Given data
k = 5      # they will sell 5 apples on a given day
mu = 3     # 3 apples per day on average

prob=round(poisson.pmf(k, mu),3)

print('The probability that the store sells 5 apples in a given day is',prob)
```

#### Question 2:A certain store sells seven footballs per day on average. What is the probability that this store sells four or less footballs in a given day?

```python
from scipy.stats import *

# Given data
k = 4      # store sells four or less footballs in a given day
mu = 7     # 3 apples per day on average

prob=round(poisson.cdf(k,mu),3)

print('The probability that the store sells four or less footballs in a given day is',prob)
```

#### Question 3: A certain store sells 15 cans of tuna per day on average. What is the probability that this store sells more than 20 cans of tuna in a given day?

```python
from scipy.stats import *

# Given data
k = 20      # store sell smore than 20 cans of tuna in a given day
mu = 15     # store sells 15 cans of tuna per day on average

prob=round(poisson.sf(k,mu),3)

print('The probability that the store sells more than 20 cans of tuna in a given day is',prob)
```


#### Question 6: Assume that the probability of an individual coalminer being killed in a mine accident during a year is 1/2400. Use Poisson distribution to calculate the probability that in a mine employing 200 miners there will be at least one fatal accident in a year.

```python
from scipy.stats import *

# Given data
n = 200        # mine employing 200 miners
p = 1/2400    # probability of a bad reaction from a certain injection is 0.001  
k = 1        # there will be at least one fatal accident in a year

prob=round(1-poisson.cdf(k-1,mu=n*p),3)

print('The probability that in a mine employing 200 miners there will be at least one fatal accident in a year',prob)
```

#### Question 7: In a certain factory turning out razor blades, there is a small chance of 0.002 for any blade to be defective. The blades are supplied in packets of 10, use Poisson distribution to calculate the approximate number of packets containing (i) no defective, (ii) one defective, (iii) two defective blades respectively in a consignment of 10,000 packets.

```python
from scipy.stats import *

# Given data
n = 10        # blades are supplied in packets of 10 in a consignment of 10,000 packets
p = 0.002    # small chance of 0.002 for any blade to be defective.
  
prob1=round(poisson.pmf(k=0,mu=n*p),4)

prob2=round(poisson.pmf(k=1,mu=n*p),4)

prob3=round(poisson.pmf(k=2,mu=n*p),4)

print('(i) Number of packets containing no defective blade is ', round(prob1*10000))

print('(ii) Number of packets containing one defective blade is',round(prob2*10000))

print('(iii) Number of packets containing two defective blades is',round(prob3*10000))
```

#### Question 8: Given that 2% of the fuses manufactured by a firm are defective. Find by using Poisson distribution, the probability that a box containing 200 fuses has (i) no defective fuses, (ii) 3 or more defective fuses, (iii) at least one defective fuse.

```python
from scipy.stats import *

# Given data
n = 200      # box containing 200 fuses
p = 0.02     # 2% of the fuses manufactured by a firm are defective

prob1=round(poisson.pmf(k=0,mu=n*p),3)
prob2=round(1-poisson.cdf(k=2,mu=n*p),3)
prob3=round(1-poisson.cdf(k=0,mu=n*p),3)

print('(i) The probability that a box containing 200 fuses has no defective fuses is ', prob1 )

print('(ii) The probability that a box containing 200 fuses has 3 or more defective fuses is',prob2)

print('(iii) The probability that a box containing 200 fuses has at least one defective fuse is',prob3)
```

#### Question 9: A certain screw making machine produces on an average two defectives out of 100 and packs them in boxes of 500.  Find by using Poisson distribution, the probability that a box containing (i) 3 defectives, (ii) at least one defective, (iii) between two and four defectives.

```python
from scipy.stats import *

# Given data
n = 500      # packs them in boxes of 500
p = 2/100     # two defectives out of 100

prob1=round(poisson.pmf(k=3,mu=n*p),4)
prob2=round(1-poisson.cdf(k=0,mu=n*p),4)
prob3=round(poisson.pmf(k=3,mu=n*p),4)

print('(i) The probability that a box containing 3 defectives is ', prob1 )

print('(ii) The probability that a box containing at least one defective is',prob2)

print('(iii) The probability that a box containing between two and four defectives is',prob3)
```

#### 10.The arrival of trucks at a receiving dock is a Poisson process with a mean arrival rate of 2 per hour. (a) Find the probability that exactly 5 trucks arrive in a two hour period. (b) Find the probability that 8 or more trucks arrive in a two hour period. (c) Find the probability that exactly two trucks arrive in a one hour period and exactly 3 trucks arrive in the next one hour period.

```python
from scipy.stats import *

# Given data
mean = 2           # mean arrival rate of 2 per hour
mu = 4            # mean for two hour period

prob1=round(poisson.pmf(k=5,mu=4),4)       # exactly 5 trucks arrive in a two hour period

prob2=round(1-poisson.cdf(k=7,mu=4),4)    # 8 or more trucks arrive in a two hour period

# exactly two trucks arrive in a one hour period and exactly 3 trucks arrive in the next one hour period
prob3=round(poisson.pmf(k=2,mu=2) * poisson.pmf(k=3,mu=2),4) 

print('(a) The probability that exactly 5 trucks arrive in a two hour period is ', prob1 )

print('(b) The probability that 8 or more trucks arrive in a two hour period is',prob2)

print('(c) The required probability is',prob3)
```

## How to Visualize a Poisson Distribution

You can visualize a poisson distribution in Python by using the matplotlib libraries:

### You can use the following syntax to plot a Poisson distribution with a given mean 3 and size = 10000

```python
from scipy.stats import *
import matplotlib.pyplot as plt

#generate Poisson distribution with sample size 10000
x = poisson.rvs(mu=3, size=10000)
print('The poisson distribution with mean 3 and size = 10000 is \n',x)
#create plot of Poisson distribution
plt.hist(x, density=True, edgecolor='black')
plt.xlabel('X axis', fontsize = 15)
plt.ylabel('Probability', fontsize = 15)
plt.xticks(range(1,15,1))
plt.title('Probability Mass Function', fontsize = 15)
plt.show()
```
