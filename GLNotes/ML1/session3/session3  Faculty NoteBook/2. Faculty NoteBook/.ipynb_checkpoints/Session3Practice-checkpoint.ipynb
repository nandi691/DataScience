```python
df = df.select_dtypes(include=[np.number])
```

```python
q1 = df.quantile(0.25,numeric_only=True)
q3 = df.quantile(0.75,numeric_only=True)

iqr = q3 - q1

outlier1 = df < ( q1 - (1.5*iqr))
outlier2 = df > ( q3 + (1.5*iqr))

df = df[~(outlier1| outlier2).any(axis=1)]
```

```python
df.shape
```

```python
plt.figure(figsize=(15,8))
df.loc[:,['Overtime', 'Other_Salaries', 'Retirement', 'Other_Benefits', 'Total_Benefits']].boxplot()

plt.title("Plot of all independent Variables after removing outliers",fontsize=15)
plt.xticks(rotation='vertical',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.boxplot(df.Total_Compensation)
plt.title("Dependent Variable \"Total Compensation\" After removing outliers",fontsize=15)
plt.xlabel("Total Compensation",fontsize=15)
plt.show()
```

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()*100).sort_values(ascending=False)

missing = pd.concat([total,percent],axis=1,keys=["Total","Percent"])
missing
```

```python
df = df.apply(lambda x:(x-x.mean())/x.std())
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.Total_Benefits,df.Total_Compensation,random_state=1,test_size=0.3)
```

```python
X_train = pd.DataFrame(X_train)
```

```python
c = np.ones(X_train.shape[0])
```

```python
X_train.insert(loc=0,column='intercept',value=c)
```

```python
X_train.head()
```

```python
import statsmodels.api as sm
```

```python
model = sm.OLS(y_train,X_train).fit()
```

```python
model.summary()
```

```python
y_train_pred = model.predict(X_train)
```

```python
y_train_pred.head()
```

```python
ssr = np.sum((y_train_pred - y_train.mean())**2)
ssr
```

```python
sse = np.sum((y_train-y_train_pred)**2)
sse
```

```python
sst = np.sum((y_train-y_train.mean())**2)
sst
```
