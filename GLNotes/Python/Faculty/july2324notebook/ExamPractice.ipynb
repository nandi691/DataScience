```python
def sales_performance_category(sales_value):
    if sales_value > 950:
        return 'Excellent_Sales'
    elif sales_value < 930:
        return 'Average_Sales'
    else:
        return 'Good_Sales'

```

```python
df['Category'] = df.Vol.map(sales_performance_category)
df
```
