```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly

import warnings 
warnings.filterwarnings('ignore')
```

```python
tips_data = sns.load_dataset('tips')
tips_data.head().transpose()
```

```python
X = np.linspace(1,10,20)
Y = np.exp(X)
plt.figure(figsize=[15,8])
plt.title("My plot study has started",fontsize=15)
plt.xlabel("X Axis",fontsize=15)
plt.ylabel("Y Axis",fontsize=15)
plt.plot(X,Y)
plt.show()
```

```python
X = np.linspace(1,10,20)
plt.figure(figsize=[15,8])
Y1 = X
Y2 = np.sqrt(X)
Y3 = X-3

plt.plot(X,Y1,'r',X,Y2,'g',X,Y3,'b')
plt.show()
```

```python
X = tips_data['total_bill']
Y = tips_data['tip']
plt.figure(figsize=[15,8])
#plt.rcParams['lines.markersize'] = np.array(Y)
plt.scatter(X,Y,s= np.array(Y)**2,c='green',alpha=0.5)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Total Bill",fontsize=15)
plt.ylabel("Tips",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=[8,6])
plt.title("Bar Plot")
tips_data.smoker.value_counts().plot(kind='bar',rot=45,color=['green','red'])
plt.xlabel("Smoker")

plt.text(x=-0.05,y=tips_data.smoker.value_counts()[0]+2,s=tips_data.smoker.value_counts()[0])
plt.text(x=1-0.03, y= tips_data.smoker.value_counts()[1]+2,s=tips_data.smoker.value_counts()[1])
plt.show()
```

```python
#plt.figure(figsize=[4,3])
#plt.title("Bar Plot")
X = tips_data.smoker.value_counts()[0]
Y = tips_data.smoker.value_counts()[1]
counts = [X,Y]
values = ["NO","YES"]
fig,ax = plt.subplots()

ax.bar(values,counts,width=0.4,color=["green","red"])
ax.set_title("Bar plot",color="brown",fontsize=15)
ax.set_ylabel("Count",fontsize=15)
ax.set_xlabel("Smoker",fontsize=15)
#ax.legend(title="Smoker")
plt.show()
```

```python
days = list(tips_data.day.unique())
value = tips_data.day.value_counts()
plt.title("Pie Plot")
plt.pie(x=value,labels=days,autopct='%.1f%%',radius=1.2,explode=[0,0.025,0,0.02])
plt.show()
```

```python
days = list(tips_data.day.unique())
value = tips_data.day.value_counts()
plt.title("Donut Plot")
plt.pie(x=value,labels=days,autopct='%.1f%%',radius=1.2,explode=[0,0.025,0,0.02])
circle= plt.Circle((0,0),radius=0.4,color='white')
plt.gcf().gca().add_artist(circle)
plt.show()
```

```python
plt.hist(tips_data.tip,bins=8)
plt.title("Histogram Plot")
plt.xlabel("Tip")
plt.ylabel("Frequency")
plt.show()
```
