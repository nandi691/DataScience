```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
name = ['Sheldon','Amy','Priya','Raj']
maths = [72,34,54,76]
geography = [98,98,78,87]
xticks = np.arange(len(name))
plt.bar(x=xticks,height=maths,width=0.4,label="Maths")
plt.bar(x=xticks+0.4,height=geography,width=0.4,label="Geography")
plt.xticks(xticks+0.2,name)
plt.xlabel("Student",fontsize=15)
plt.ylabel("Marks",fontsize=15)
plt.title("Marks of Students")
plt.legend()
plt.show()
```

```python
# horizontal barplot   
name = ['Sheldon','Amy','Priya','Raj']
maths = [72,34,54,76]
geography = [98,98,78,87]
yticks = np.arange(len(name))
plt.figure(figsize=(15,8))
plt.barh(y=yticks,width=maths,label="Maths",height=0.4)
plt.barh(y=yticks+0.4,width=geography,label="Geography",height=0.4)
plt.yticks(yticks+0.2,name)
plt.ylabel("Student",fontsize=15)
plt.xlabel("Marks",fontsize=15)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# creating the dataset
data = {'C': 20, 'C++': 15, 'Java': 30,
		'Python': 35}

courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize=(10, 5))

# creating the bar plot
plt.barh(courses, values, color='maroon')

plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```
