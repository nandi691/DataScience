```python
sales = {'Months': ['Jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec'],
        'Sales': [22000,27000,25000,29000,35000,67000,78000,67000,56000,56000,89000,60000],
         'Seasons': ['Winter','Winter','Spring','Spring','Spring','Summer','Summer','Summer','Fall','Fall','Fall','Winter']
        }

df = pd.DataFrame(sales,columns=['Months','Sales','Seasons'])
df
```

```python
pd.pivot_table(df,index=["Months", "Seasons"])
```

```python
pd.pivot_table(df,index=["Seasons"],values=["Sales"])
```

```python
pd.pivot_table(df,index=["Seasons"],values=["Sales"],aggfunc=np.std)
```

```python
pd.pivot_table(df,index=["Seasons"],values=["Sales"],aggfunc=[np.std,len])
```

```python
pd.pivot_table(df,index=["Months"],values=["Sales"],columns=["Seasons"],aggfunc=[np.sum])
```

```python
pd.pivot_table(df,index=["Months"],values=["Sales"],columns=["Seasons"],aggfunc=[np.sum],fill_value=0)
```
