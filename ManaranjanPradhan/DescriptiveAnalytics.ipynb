```python
soldprice_by_age = ipl_auction_df.groupby('AGE')['SOLD PRICE'].mean().reset_index()
soldprice_by_age
```

```python
ipl_auction_df.groupby(['AGE','PLAYING ROLE'])['SOLD PRICE'].mean()
```

```python
soldprice_by_age_role = ipl_auction_df.groupby(['AGE','PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
soldprice_by_age_role
```

```python
soldprice_comparison = soldprice_by_age_role.merge(soldprice_by_age,on='AGE',how='outer')
soldprice_comparison
```

```python
soldprice_comparison.rename(columns={
    'SOLD PRICE_x':'SOLD PRICE ROLE',
    'SOLD PRICE_y':'SOLD PRICE AGE'
},inplace=True)
soldprice_comparison
```

```python
soldprice_comparison['change'] = soldprice_comparison.apply(lambda x:(x['SOLD PRICE ROLE']-x['SOLD PRICE AGE'])/x['SOLD PRICE AGE'],axis=1)
```
